```python
x = galaxy['INTERNET']
y = galaxy['EQUIPO_COMPUTO']
galaxy.head()
```

```python
#
# 6. select
#

from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
```

```python
#
# 6. try some other parameters
#

model = LinearRegression(fit_intercept=True)
model
```

```python
print(x)
x.shape
```

```python
#
# 2. prep (etl)
#

X = x[:, np.newaxis]
X.shape
```

```python
#
# 6. train
#

model.fit(X, y)
```

```python
#
# 7. evaluate (x)
#

model.coef_
```

```python
#
# 7. evaluate (y)
#

model.intercept_
```

```python
#
# 2. prep (etl)
#

xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
print(xfit.shape)
Xfit.shape
```

```python
yfit = model.predict(Xfit)
```

```python
#
#  4. eda (again)
#

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.show()
```
